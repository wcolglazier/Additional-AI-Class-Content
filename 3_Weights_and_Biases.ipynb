Neural networks are a foundational concept in AI that can identify trends or patterns in data that may be invisible to the human eye. The goal of a neural network is to process input data and transform it into meaningful output. This transformation occurs through layers of neurons, which apply mathematical operations to the inputs.

Architecture.jpg.webp

The goal of a neural network is to transform raw input data into meaningful information by applying weights and biases. These parameters help the network detect patterns, make predictions, and classify data by refining it through multiple layers. Weights determine the influence of each input on the output, with larger values having a greater impact and smaller values contributing less. To better understand how neural networks work, we often use a 'toy dataset'â€”a simple, artificial dataset designed for testing and learning purposes. While it doesn't reflect real-world data, it helps in exploring the network's functionality.

```python
import numpy as np

Inputs = [
     [1, 2, 3, 2.5],          # Input 1
     [2.0, 5.0, -1.0, 2.0],   # Input 2
     [-1.5, 2.7, 3.3, -0.8]   # Input 3
]

Weights = [
    [0.2, 0.8, -0.5, 1.0],
    [0.5, -0.91, 0.26, -0.5],
    [-0.26, -0.27, 0.17, 0.87]
]

Biases = [2.0, 3.0, 0.5]


Output = np.dot(Inputs, np.array(Weights).T) + Biases


print(Output)

```

```python
import numpy as np

Data = [[3.0, -2.0, 2.2],
        [2.0, -5.0, -8.0],
        [-1.5, 2.7, 3.3]]

print(np.array(Data))

Output_With_ReLU = np.maximum(0, Data)

print("\nOur data with a ReLU function")
print(Output_With_ReLU)
```

```python
import numpy as np

Inputs = [
     [1, 2, 3, 2.5],          # Input 1
     [2.0, -5.0, -1.0, 2.0],   # Input 2
     [-1.5, 2.7, 3.3, -0.8]   # Input 3
]

Weights = [
    [0.2, 0.8, -0.5, 1.0],
    [0.5, -0.91, 0.26, -0.5],
    [-0.26, -0.27, 0.17, 0.87]
]

Biases = [2.0, -3.0, 0.5]

Output = np.dot(Inputs, np.array(Weights).T) + Biases

Output_With_ReLU = np.maximum(0, Output)


print("Output before ReLU")
print(Output)

print("\nOutput after ReLU")
print(Output_With_ReLU)
```

```python
import numpy as np

inputs = [[1, 2, 3, 2.5],
     [2.0, 5.0, -1.0, 2.0],
     [-1.5, 2.7, 3.3, -0.8]]

weights1 = [[0.2, 0.8, -0.5, 1.0],
           [0.5, -0.91, 0.26, -0.5],
           [-0.26, -0.27, 0.17, 0.87]]

biases1 = [2.0, 3.0, 0.5]

weights2 = [[0.1, -0.14, 0.5],
            [-0.5, 0.12, -0.33],
            [-0.44, 0.73, -0.13]]

biases2 = [-1, 2, -0.5]


layer1_output = np.dot(inputs, np.array(weights1).T) +biases1

layer2_output = np.dot(layer1_output, np.array(weights2).T) +biases2


print(layer2_output)
```
