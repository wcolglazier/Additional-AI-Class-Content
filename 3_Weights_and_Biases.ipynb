Neural networks are a foundational concept in AI that can identify trends or patterns in data that may be invisible to the human eye. The goal of a neural network is to process input data and transform it into meaningful output. This transformation occurs through layers of neurons, which apply mathematical operations to the inputs.

Architecture.jpg.webp

The goal of a neural network is to transform raw input data into meaningful information by applying weights and biases. These parameters help the network detect patterns, make predictions, and classify data by refining it through multiple layers. Weights determine the influence of each input on the output, with larger values having a greater impact and smaller values contributing less. To better understand how neural networks work, we often use a 'toy dataset' a simple, artificial dataset designed for testing and learning purposes. While it doesn't reflect real-world data, it helps in exploring the network's functionality.

```python
import numpy as np

Inputs = [
     [1.0, 2.2, 3.1, 2.5],    # Input 1
     [2.0, 5.0, -1.0, 2.0],   # Input 2
     [-1.5, 2.7, 3.3, -0.8]   # Input 3
]

Weights = [
    [0.2, 0.8, -0.5, 1.0],
    [0.5, -0.91, 0.26, -0.5],
    [-0.26, -0.27, 0.17, 0.87]
]

Biases = [2.0, 3.0, 0.5]


Output = np.dot(Inputs, np.array(Weights).T) + Biases


print(Output)

```

Activation functions, like ReLU (Rectified Linear Unit), transform data by setting negative values to zero while keeping positives unchanged. This introduces non-linearity, helping the network focus on key features.

[ReLU Video](https://www.youtube.com/watch?v=68BZ5f7P94E)

ReLU-activation-function-new.png

We can apply a ReLU function to our dataset by setting all negative values to zero.

```python
import numpy as np

Data = [[3.0, -2.0, 2.2],
        [2.0, -5.0, -8.0],
        [-1.5, 2.7, 3.3]]

print("Output before ReLU")
print(np.array(Data))

Output_With_ReLU = np.maximum(0, Data)

print("\nOutput after ReLU")
print(Output_With_ReLU)
```

We can now add the ReLU function to our original neural network.

```python
import numpy as np

Inputs = [
     [1.0, 2.2, 3.1, 2.5],    # Input 1
     [2.0, 5.0, -1.0, 2.0],   # Input 2
     [-1.5, 2.7, 3.3, -0.8]   # Input 3
]

Weights = [
    [0.2, 0.8, -0.5, 1.0],
    [0.5, -0.91, 0.26, -0.5],
    [-0.26, -0.27, 0.17, 0.87]
]

Biases = [2.0, 3.0, -1.5]

Output = np.dot(Inputs, np.array(Weights).T) + Biases

Output_With_ReLU = np.maximum(0, Output)


print("Output before ReLU")
print(Output)

print("\nOutput after ReLU")
print(Output_With_ReLU)
```

Now we can create a more realistic neural network with four layers of weights and biases that influence the inputs.

```python
import numpy as np
import numpy as np

Inputs = [[1, 2, 3, 2.5],
          [2.0, 5.0, -1.0, 2.0],
          [-1.5, 2.7, 3.3, 0.8]]

Weights1 = [[0.2, 0.8, 0.5, 1.0],
            [0.5, -0.91, 0.26, -0.5],
            [-0.26, 0.27, 0.17, 0.87]]
Biases1 = [2.0, 3.0, 0.5]

Weights2 = [[0.1, -0.14, 0.5],
            [0.5, 0.12, -0.33],
            [0.44, 0.73, -0.13]]
Biases2 = [-1, 2, -0.5]

Weights3 = [[0.3, -0.2, 0.8],
            [-0.7, 0.15, 0.2],
            [0.5, -0.3, -0.1]]
Biases3 = [0.5, 1.5, 1.0]

Weights4 = [[-0.4, 0.6, -0.2],
            [0.8, 0.3, 0.4],
            [-0.1, 0.9, -0.5]]
Biases4 = [1.0, 0.5, 0.2]

Output1 = np.dot(Inputs, np.array(Weights1).T) + Biases1
Output1_With_ReLU = np.maximum(0, Output1)

Output2 = np.dot(Output1_With_ReLU, np.array(Weights2).T) + Biases2
Output2_With_ReLU = np.maximum(0, Output2)

Output3 = np.dot(Output2_With_ReLU, np.array(Weights3).T) + Biases3
Output3_With_ReLU = np.maximum(0, Output3)

Output4 = np.dot(Output3_With_ReLU, np.array(Weights4).T) + Biases4
Final_Output = np.maximum(0, Output4)


print(Final_Output)
```
